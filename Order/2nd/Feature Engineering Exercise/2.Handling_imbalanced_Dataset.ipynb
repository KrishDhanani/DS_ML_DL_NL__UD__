### Handling imbalanced Dataset 
-> In our Dataset if there are 1000 samples and dataset has categorical values like 'Yes' and 'No'. 
-> Now the main thing is; if in this Dataset the 900 are 'Yes' and only 100 are 'No' Data and on this Data if we train our model then our model prediction go for biased decision.
-> Either we have to decrease 900yes data or we can improve 100no data; so that our model prediction is no more biased base. 
-> For that we use "Up scaling" and "Down Scaling".

## Handling Imbalanced Dataset

1. Up Sampling
2. Down Sampling

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123) # set the seed

# print(np.random.randint(100))  # bcs we set the seed it always returns the same number

# creating dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
# Creating DataFrame with Imbalanced data-set

# loc = Mean (“centre”) of the distribution.
# scale = Standard deviation (spread or “width”) of the distribution. Must be non-negative.

class_0 = pd.DataFrame({
    'feature-1': np.random.normal(loc=0, scale=1, size=n_class_0), 
    'feature-2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature-1': np.random.normal(loc=0, scale=1, size=n_class_1),
    'feature-2': np.random.normal(loc=0, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

## UP-SAMPLING

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
from sklearn.utils import resample

df_minority_upsampled = resample(df_minority, replace=True, # Sample with replacement
                                n_samples = len(df_majority),
                                 random_state = 42    # setting state gives us the best same output as we set seed 
                                )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled
```

```python
df_upsampled['target'].value_counts()
```

## DOWN-SAMPLING

```python
import pandas as pd

# Setting random seed for reproductivity
np.random.seed(123)

# create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
# Making DataFrames

class_0 = pd.DataFrame({
    'feature-1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature-2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})
class_1 = pd.DataFrame({
    'feature-1': np.random.normal(loc=0, scale=1, size=n_class_1),
    'feature-2': np.random.normal(loc=0, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

```python
## downsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample

df_majority_downsampled = resample(df_majority, 
                                  n_samples = len(df_minority),
                                  random_state = 42)
```

```python
df_majority_downsampled.shape
```

```python
df_majority_downsampled.head()
```

```python
df_downsampled = pd.concat([df_minority, df_majority_downsampled])
```

```python
df_downsampled['target'].value_counts()
```
