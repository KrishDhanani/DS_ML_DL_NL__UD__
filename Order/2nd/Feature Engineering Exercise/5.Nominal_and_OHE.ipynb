So what is Encoding:

2. Now on the 'Degree' feature contains categorical values like: B.E., PHD, Masters etc. and on the basis of this data you want to predict the salary.
3. For that our ML model can only understand numerical data and converting these categorical data into numerical one called Encoding.


### Nominal/OHE Encoding
One hot encoding, also known as nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green, blue), we can represent it using one hot encoding as follows:

1. Red: [1, 0, 0]
2. Green: [0, 1, 0]
3. Blue: [0, 0, 1]

## Disadvantage

1. If in Feature there are 100 diffrent categories then it has to create for all 100 categories different different column and then assign 1 and 0 to relevent feature.
2. At the end you can see our Data look like "Sparce metrix" which leads to "Overfitting Problem".

## Sparse & Dense Metrix
 A sparse matrix or sparse array is a matrix in which most of the elements are zero. There is no strict definition regarding the proportion of zero-value elements for a matrix to qualify as **sparse** but a common criterion is that the number of non-zero elements is roughly equal to the number of rows or columns. By contrast, if most of the elements are non-zero, the matrix is considered **dense**.

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create simple DataFrame
df = pd.DataFrame({
    'color': ['red','blue','green','green','red','blue']
})
df.head()
```

```python
# Create instance of OneHotEncode
encoder = OneHotEncoder()
```

```python
# Now let's perform fit and then transform
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded   # Here you can see the red goes to ende bcs it sorts all data into alphabetically
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoder_df      # Now i can send my entire data for training model
```

```python
# So now what if we have NEW DATA...
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df, encoder_df], axis=1)
```
