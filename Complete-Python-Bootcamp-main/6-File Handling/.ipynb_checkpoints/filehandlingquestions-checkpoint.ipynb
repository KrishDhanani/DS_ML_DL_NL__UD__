```python
#Assignment 6:
original_word = 'Falcon'
new_word = 'funny'
with open('log.txt', 'r') as f:
    text = f.read()
    print(type(text))
new_text = text.replace(original_word, new_word)
with open('log.txt', 'w') as f:
    f.write(new_text)
    
```

```python
# Assignment 7:

with open('log.txt', 'r') as file:
    original_list = file.readlines()
    reversed_list = original_list[::-1]
    for line in reversed_list:
        print(line.strip())
```

```python
# Assignment 8:

with open('log.txt', 'r') as file:
    list_ = file.readlines()
    lines_count = len(list_)
    word_count = 0
    char_count = 0

    for line in list_: 
        word_count += len(line.split())
        for ch in line.split():
            char_count += len(ch)
            
    print(word_count)
    print(char_count)
```

```python
# Assignment 9:

with open('example.txt', 'r') as ex_f:
    with open('document.txt', 'r') as do_f:
        with open('merged.txt', 'w') as me_f:
            whole_content = ex_f.read() + do_f.read()
            print(whole_content) 
            me_f.write(whole_content)
```

```python
# Assignment 10:
import datetime
def log_message(message, filename='activity.log'):
    timestamp = datetime.datetime.now().isoformat()
    with open(filename, 'a') as file:
        file.write(f'[{timestamp}] {message}\n')

print(log_message('This is the msg.'))
```
