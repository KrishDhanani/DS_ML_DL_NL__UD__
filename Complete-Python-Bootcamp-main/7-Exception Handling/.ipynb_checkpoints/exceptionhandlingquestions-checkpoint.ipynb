```python
# Assignment 1:

def fun(a, b):
    try:
        division = a/b
    except ZeroDivisionError:
        print("Enter the valid devidor(b)")
    finally:
        print("Code runed successfully")

fun(1,0)
```

```python
# Assignment 2:

def read_file():
    try:
        with open('data.txt', 'r') as f:
            content = f.read()
            print(content)
    except FileNotFoundError as fno:
        print(fno)

read_file()
```

```python
# Assignment 8:

import requests 

def content(url):
    try:
        response = requests.get(url)
        print(response.status_code)
        print(response.text)
    except requests.RequestException as e:
        print(e)
    finally:
        print("Execution Commplete")

# content('https://jsonplaceholder.typicode.com/posts/1')
content('https://nonexistent.url')
```

```python
# Assignment 10:
lst = [1,1,-1]

class NegativeNumberError(Exception):
    pass

def check_negativ(l):
    try:
        for num in l:
            if num < 0:
                raise NegativeNumberError('In list negative number Present.')
    except NegativeNumberError as e:
        print(e)
    finally:
        print('Execution Complete.')

check_negativ(lst)
```

```python
# Assignment 11:
def risky_function():
    raise ValueError("An error occurred in risky_function.")

def safe_function():
    try:
        risky_function()
    except ValueError as e:
        print(f"Error: {e}")
    finally:
        print("Execution complete.")

safe_function()
```

```python
# Assignment 12:

class operation:

    def division(self,a,b):
        try:
            d = a/b
            return d
        except ZeroDivisionError as e:
            print(e)

op = operation()
op.division(2,4)
```

```python
# Assignment 13:

lst = ['k', 'd', 9]

def convert_str_to_int(l):
    try:
        for value in l:
            int(value)
    except ValueError as e:
        print(e)

convert_str_to_int(lst)
```

```python
# Assignment 14:
def convert_with_comprehension(lst):
    try:
        integers = [int(item) for item in lst]
    except ValueError as e:
        print(f"Error: {e}")
        integers = None
    finally:
        print("Execution complete.")
    return integers
convert_with_comprehension(['1', '2', 'three', '4'])
```

```python
# Assignment 15:

def write_strings_to_file(strings, filename):
    try:
        file = open(filename, 'w')
        for string in strings:
            file.write(string + '\n')
    except IOError as e:
        print(f"Error: {e}")
    finally:
        try:
            file.close()
        except NameError:
            pass

# Test
# write_strings_to_file(['Hello', 'World'], 'output.txt')
```
