```python
# Assignment 9

import random

def get_number():
    num = random.randint(1,10)
    return num

matrix = []

temp_lst = []
for i in range(3):
    for i in range(3):
        temp_lst.append(get_number())
    matrix.append(temp_lst)
    temp_lst = []

matrix_2 = []

i = 0
for j in range(len(matrix)):
    for index in range(len(matrix)):
        temp_lst.append(matrix[index][i])
    matrix_2.append(temp_lst)
    temp_lst = []
    i+=1

print(matrix)
print(matrix_2)
```
