**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
n = int(input())
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(n))
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year = 2026
if year%4==0:
    if year%100==0:
        if year%400==0:
            print("Is leap year")
        else:
            print("Not leap year")
    else:
        print("Is leap year")
else:
    print("Not leap year")
```
