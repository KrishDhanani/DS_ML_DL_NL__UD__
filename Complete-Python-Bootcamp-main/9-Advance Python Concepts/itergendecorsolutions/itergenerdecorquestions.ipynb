Write three generator functions: `integers` that yields integers from 1 to 10, `doubles` that yields each integer doubled, and `negatives` that yields the negative of each doubled value. Chain these generators to create a pipeline that produces the negative doubled values of integers from 1 to 10.

```python
# Assignment 1:

class Countdown():
    def __init__(self, number):
        self.current = number

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= 0:
            raise StopIteration

        else:
            self.current -= 1
            return self.current

for number in Countdown(5):
    print(number)
```

```python
# Assignment 2:
class MyRange:
    def __init__(self, number):
        self.current = -1
        self.max = number

    def __iter__(self):
        return self

    def __next__(self):
        if self.max-1 == self.current:
            raise StopIteration

        else:
            self.current += 1
            return self.current

for i in MyRange(3):
    print(i)
```

```python
# Assignment 3:
def fibonacci(n):
    n1 = 0 
    yield n1
    n2 = 1
    for i in range(n):
        n3 = n1 + n2
        n1 = n2
        n2 = n3
        yield n3

f = fibonacci(10)
```

```python
# Assignment 5:
def even_numbers(n):
    for i in range(n+1):
        if i % 2 == 0:
            yield i

def squares(n):
    for i in n:
        yield i**2

even_gen = even_numbers(20)
square_gen = squares(even_gen)
for square in square_gen:
    print(square)
```
