```python
# Assignment 1:

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

class dog(Animal):
        def __init__(self, name, species, breed):
            super().__init__(name, species)
            self.breed = breed

d = dog("sultan", "hybrid", "semihuman")
print(d.name)
print(d.species)
            
```

```python
# Assignment 2:

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

class dog(Animal):
        def __init__(self, name, species, breed):
            super().__init__(name, species)
            self.breed = breed

        def __str__(self):
            return f"My dog name is {self.name}"
            
d = dog('Buddy', 'Canine', 'Golden Retriever')
print(d)
```

```python
# Assignment 3:

class Dog:
    def __init__(self):
        pass

    def bark(self):
        return "Woof! Woof!"

d = Dog()
print(d.bark())
```

```python
# Assignment 4:

class Walker:
    def __init__(self):
        pass
        
    def walk(self):
        print("Hie, I'm walking with my dog.")

class Runner:
    def __init__(self):
        pass

    def run(self):
        print("Run...Run...Run...")

class Athlete(Walker, Runner):
    def __init__(self):
        Walker.__init__(self)
        Runner.__init__(self)
        Walker.walk(self)
        Runner.run(self)

a = Athlete()
```

```python
# Assignment 5:

class Athlete1(Walker, Runner):
    def walk(self):
        super().walk()
        print("I'm walking alone...")

a1 = Athlete1()
a1.walk()
```

```python
# assignment 6:
class Athlete2(Athlete1):
    def __init__(self, time):
        self.training_hour = time

    def train(self):
        print(f"Train your self atleast {self.training_hour} hours")

a2 = Athlete2(1)
a2.train()
```

```python
# Assignment 7:

class A:
    def show(self):
        print("This is class A.")

class B(A):
    def show(self):
        print("This is class B.")


class C(A):
    def show(self):
        print("This is class C.")

class D(B, C):
    pass
    
d = D()
d.show()
```

```python
# Assignment 8:

class Shape:
    def __init__(self, color):
        self.color = color

class Circle(Shape):
    def __init__(self, color, radious):
        self.radious = radious
        super().__init__(color)

c = Circle("red", 23)
print(c.radious)
```

```python
# Assignment 9:

class Person:
    def __init__(self, name):
        self.name = name

class Employee:
    def __init__(self, e_id):
        self.e_id = e_id

class Manager(Person, Employee):
    def __init__(self, name, e_id):
        Person.__init__(self, name)
        Employee.__init__(self, e_id)

m = Manager("Max", 33)
print(m.name)
    
```

```python
# Assignment 10:
class Vehicle:
    def start(self):
        print("Vehicle has been started...")

class Car(Vehicle):
    def start(self):
        super().start()
        print("Car starting...")

c = Car()
c.start()
```

```python
# Assignment 11:

class Flyer:
    def fly(self):
        print("Fly...Fly...")

class Swimmer:
    def swim(self):
        print("Swim...Swim...")

class Superhero(Flyer, Swimmer):
    pass

sh = Superhero()
sh.fly()
sh.swim()
```

```python
# Assignment 12:

class Base1:
    def __init__(self, a):
        self.a = a

class Base2:
    def __init__(self, b):
        self.b = b

class Derived(Base1, Base2):
    def __init__(self, a, b, c):
        Base1.__init__(self, a)
        Base2.__init__(self, b)
        self.c = c

d = Derived(1, 2, 3)
print(d.a)
print(d.b)
print(d.c)
```

```python
# Assignment 13:

class Animal:
    pass

class Cat(Animal):
    pass

a = Animal()
c = Cat()

print(isinstance(a, Animal))
print(isinstance(c, Cat))
print(isinstance(c, Animal))
print(isinstance(a, Cat))
```

```python
# Assignment 14:

class Bird:
    def speak(self):
        pass

class Parrot(Bird):
    def speak(self):
        print("Parrot says: Squawk!")

class Penguin(Bird):
    def speak(self):
        print("Penguin says: Honk!")

# Test
birds = [Parrot(), Penguin()]
for bird in birds:
    bird.speak()

```

```python
# Assignment 15:

class Device:
    def __init__(self, brand):
        self.brand = brand

class Phone(Device):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

class Camera:
    def __init__(self, resolution):
        self.resolution = resolution

class SmartPhone(Phone, Camera):
    def __init__(self, brand, model, resolution):
        Phone.__init__(self, brand, model)
        Camera.__init__(self, resolution)

smartphone = SmartPhone('Apple', 'iPhone 12', '12 MP')
print(smartphone.brand, smartphone.model, smartphone.resolution)
```
