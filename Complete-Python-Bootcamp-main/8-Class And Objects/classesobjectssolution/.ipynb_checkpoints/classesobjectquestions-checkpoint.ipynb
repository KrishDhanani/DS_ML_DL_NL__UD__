```python
# Assignment 4:

class BankAccount():

    def __init__(self, accountNumber, balance):
        self.accountNumber = accountNumber
        self.balance = balance

    def get_accountNumber(self):
        return self.accountNumber
    
    def get_balance(self):
        return self.balance
    
    def deposit(self, amount):
        self.balance += self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            print("Entered amount is more than existing balance")
        else:
            self.balance -= amount

# Test
account = BankAccount('12345678', 1000)
account.deposit(500)
account.withdraw(200)
print(account.get_balance())  # 1300
account.withdraw(2000)  # Insufficient balance!
```

```python
# Assignment 5:

# Base class
class Person():
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Sub class
class Employee(Person):

    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

e = Employee('raj', 20, 123)
print(e.name)
```

```python
# Assignment 7:

class Address:
    def __init__(self, street, city, zipcode):
        self.street = street
        self.city = city
        self.zipcode = zipcode

class Person:
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address

# Test
address = Address('123 Main St', 'New York', '10001')
person = Person('John', 25, address)
print(person.address.street, person.address.city, person.address.zipcode)
    
```

```python
# Assignment 8:

class Counter:
    count = 0

    def __init__(self):
        Counter.count += 1

    @classmethod
    def get_count(cls):
        return cls.count

# Test
c1 = Counter()
c2 = Counter()
c3 = Counter()
print(Counter.get_count())  # 3
```

```python
# Assignment 9:
import math

class MathOperations:
    
    @staticmethod
    def s_of_num(num):
        return math.sqrt(num)

print(MathOperations.s_of_num(5))
```

```python
# Assignment 10:

class Rectangle:

    def __init__(self, length, width):
        self.__length = length
        self.__width = width

    # length property
    @property
    def length(self):
        return self.__length

    @length.setter
    def length(self, value):
        self.__length = value

    # width property
    @property
    def width(self):
        return self.__width

    @width.setter
    def width(self, value):
        self.__width = value


# Test
r = Rectangle(10, 20)
print(r.length)   # 10
print(r.width)    # 20

r.length = 50
r.width = 100
print(r.length)   # 50
print(r.width)    # 100
```

```python
# Assignment 11:
from abc import ABC, abstractmethod

class Shape:

    @abstractmethod
    def area():
        pass

class Circle(Shape):

    def __init__(self, radious):
        self.radious = radious

    def area(self):
        return 3.14*self.radious*self.radious

class Square(Shape):

    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length*self.length

circle = Circle(5)
square = Square(4)
print(circle.area())  # 78.53981633974483
print(square.area())  # 16
```

```python
# Assignment 12:

class Vector:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2
print(v3)  # Vector(6, 8)
```

```python
# Assignment 13:
class InsufficientBalanceError(Exception):
    pass

class BankAccount:

    def __init__(self):
        self.balance=1000

    def widrawl(self, amount):
        try:
            if amount > self.balance:
                raise InsufficientBalanceError("You request more amount than existing balance.")
            else:
                self.balance -= amount
                print(f"Your remaining balance:{self.balance}")
        except Exception as e:
            print(e)

a = BankAccount()
a.widrawl(1200)
```

```python
# Assignment 14:
class FileManager:
    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file

    def __exit__(self, exc_type, exc_value, traceback):
        self.file.close()

# Test
with FileManager('sample.txt', 'r') as file:
    content = file.read()
    print(content)
            
```

```python
# Assignment 15:

class Calculator:
    def __init__(self, value=0):
        self.value = value

    def add(self, amount):
        self.value += amount
        return self

    def subtract(self, amount):
        self.value -= amount
        return self

    def multiply(self, amount):
        self.value *= amount
        return self

    def divide(self, amount):
        if amount != 0:
            self.value /= amount
        else:
            print("Cannot divide by zero!")
        return self
        
calc = Calculator()
calc.add(10).subtract(3).multiply(2).divide(2)
print(calc.value)  # 7.0
    
```
