```python
# Assignment 1:
import math

class Shape:
    def area(self):
        pass

class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius

class Square(Shape):

    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length ** 2
        
Shapes = [Circle(10), Square(2)]
for shape in Shapes:
    print(shape.area())
```

```python
# Asignment 2:

def describe_shape(shape):
    print(f"The area of shape: {shape.area()}")

for shape in Shapes:
    describe_shape(shape)
```

```python
# Assignment 3:
from abc import ABC, abstractmethod

class Vehicle:

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car is starting...")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike is starting...")

c = Car()
c.start_engine()
b = Bike()
b.start_engine()
```

```python
# Assignment 4:

class Vehicle:

    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def fuel_type(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car is starting...")

    def fuel_type(self):
        print("Car running on Diesel.")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike is starting...")

    def fuel_type(self):
        print("Bike running on Petrol.")

c = Car()
c.fuel_type()
b = Bike()
b.fuel_type()
```

```python
# Assignment 5:

class BankAccount:
    def __init__(self, acc_no):
        self.__account_number = acc_no
        self.__balance = 1000

    def deposit(self, amount):
        self.__balance += amount

    def widrawl(self, amount):
        if amount > self.__balance:
            print("Entered amount is more than existing balance.")
        else:
            self.__balance -= amount 

    def check_balance(self):
        return self.__balance

ba = BankAccount(123)
ba.deposit(100)
ba.widrawl(500)
print(ba.check_balance())
```

```python
# Assignment 7:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def set_name(self, name):
        self.name = name
        
    def get_name(self):
        return self.name

    def set_age(self, age):
        self.age = age

    def get_age(self):
        return self.age

class Student(Person):
    def __init__(self, name, age, s_id):
        super().__init__(name, age)
        self.student_id = s_id

student = Student('John', 20, 'S123')
print(student.get_name(), student.get_age(), student.student_id)
student.set_name('Alice')
student.set_age(22)
print(student.get_name(), student.get_age(), student.student_id)
```
