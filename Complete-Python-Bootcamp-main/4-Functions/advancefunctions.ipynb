```python
# Assignment 3:
def get_only_integer(**kwargs):
    dict = {}
    for key,value in kwargs.items():
        if type(value) == int:
            dict[key] = value
    return dict


print(get_only_integer(a=1, b='two', c=3, d=4.5))  # {'a': 1, 'c': 3}
print(get_only_integer(x=10, y='yes', z=20))  # {'x': 10, 'z': 20}
```

```python
# Assignment 4:
def apply_callback(callaback, lst):
    return [callaback(x) for x in lst]

# Test
print(apply_callback(lambda x: x**2, [1, 2, 3, 4]))  # [1, 4, 9, 16]
print(apply_callback(lambda x: x+1, [1, 2, 3, 4]))  # [2, 3, 4, 5]
```

```python
# Assignment 7:

def higher_order_fun(filter_fun, map_fun, lst):
    return list(map_fun(i) for i in lst if filter_fun(i))

# test
print(higher_order_fun(lambda x:x>5, lambda x:x-1, [1,2,3,4,5]))
print(higher_order_fun(lambda x:x<5, lambda x:x+4, [0,2,3,4,5,6]))
```
